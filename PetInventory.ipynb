```python
class  Pet:
    def __init__(self, petName, dateOfBirth, birthWeight, owner):
        self.petName = petName
        self.dateOfBirth = dateOfBirth
        self.birthWeight = birthWeight
        self.owner = owner
    def computeDays(self, dateOfBirth):
#         currDate = date(2023,5,16)
        currDate = date.today()
        numOfDays = (currDate - self.dateOfBirth).days
        return numOfDays
        
class Owner:
    def __init__(self, petOwner, petOwnerAddress):
        self.petOwner = petOwner
        self.petOwnerAddress = petOwnerAddress
    def __str__(self):
        return self.petOwner + " lives in " + self.petOwnerAddress
        
class Mammal(Pet):
    def __init__(self, petName, dateOfBirth, birthWeight, owner, litterSize, hasClaws):
        super().__init__(petName, dateOfBirth, birthWeight, owner)
        self.litterSize = litterSize
        self.hasClaws = hasClaws
    
    def computeCurrentWeight(self):
        numOfDays = self.computeDays(self.dateOfBirth)
        bw_pounds = self.birthWeight/16
        if numOfDays > 300:
            bw_pounds *= 1.1 ** 5
        else:
            p = numOfDays//60
            bw_pounds *= 1.1 ** p
            
        return f"{self.petName}'s cur_weight in pounds is {bw_pounds:.2f}"
    
class Fish(Pet):
    def __init__(self, petName, dateOfBirth, birthWeight, owner, scaleCondition, length):
        super().__init__(petName, dateOfBirth, birthWeight, owner)
        self.scaleCondition = scaleCondition
        self.length = length
    #Weight increases by 5% every 80 days for the first 240 days and then stays constant. Assume that the increase occurs on each 80th day only.
    def computeCurrentWeight(self):
        numOfDays = self.computeDays(self.dateOfBirth)
        bw_pounds = self.birthWeight/16
        if numOfDays > 240:
            bw_pounds *= 1.05 ** 3
        else:
            p = numOfDays//80
            bw_pounds *= 1.05 ** p

        return f"{self.petName}'s cur_weight in pounds is {bw_pounds:.2f}"

class Amphibian(Pet):
    def __init__(self, petName, dateOfBirth, birthWeight, owner, numberOfLimbs, isVenomous):
        super().__init__(petName, dateOfBirth, birthWeight, owner)
        self.numberOfLimbs = numberOfLimbs
        self.isVenomous = isVenomous
        
    #Weight increases by 5% every 120 days for the first 360 days and then increases by 3% every 120 days. Assume that the increase occurs on each 120th day only.
    def computeCurrentWeight(self):
        numOfDays = self.computeDays(self.dateOfBirth)
        bw_pounds = self.birthWeight/16
        if numOfDays < 360:
            p = numOfDays//120
            bw_pounds *= 1.05 ** p
        else:
            t = (numOfDays - 360)//120
            bw_pounds *= (1.05 ** 3) * (1.03 ** t)
        
        return f"{self.petName}'s cur_weight in pounds is {bw_pounds:.2f}"
  
    
```

```python
import pickle
def read_file():
    pet_list = []
    try:
        file = open('petdata.dat', 'rb')
    except FileNotFoundError:
        return pet_list
    
    else:
        try:
            while(True):
                pet_list.append(pickle.load(file))
                
        except EOFError:
            file.close()
            return pet_list
```

```python
from datetime import date
def get_animal_input():
    err = True
    while err:
        try:
            pet_name = input("Enter the pet's name: ")
            pet_birthyear = int(input("Enter the pet's year of birth: "))
            pet_birthmonth = int(input("Enter the pet's month of birth: "))
            pet_birthday = int(input("Enter the pet's day of birth: "))
            pet_DOB = date(pet_birthyear,pet_birthmonth,pet_birthday)
            pet_weight = int(input("Enter the birth weight for the pet in ounces: "))
            if pet_weight < 0:
                raise ValueError
        except ValueError as ve:
            print("Enter positive integer", ve)
        except Exception as e:
            print("Enter valid input", e)
        else:
            err = False
    return (pet_name, pet_DOB, pet_weight)

def get_owner_input():
    pet_owner = input("Enter owner name: ")
    pet_owneraddress = input("Enter owner address: ")
    
    return (pet_owner, pet_owneraddress)

def get_mammal_input():
    err = True
    while err:
        try:
            lSize = int(input("Enter litter size: "))
            if lSize < 0:
                raise ValueError
        except ValueError:
            print("Enter positive integer")
        else:
            err = False
    err = True
    while err:
        try:
            hClaws = input("Does the pet has claws? Enter 'y' or 'n': ").lower()
            if hClaws != 'y' or hClaws != 'n':
                raise ValueError
        except ValueError:
            print('Enter y or n!')
        else:
            err = False
    
    return (lSize, hClaws)

def get_amphibian_input():
    err = True
    while err:
        try:
            limbCount = int(input("Enter the number of limbs for the pet: "))
            if limbCount < 0:
                raise ValueError
        except ValueError:
            print("Enter positive integer")
        else:
            err = False
    err = True
    while err:
        try:
            venomous = input("Is the pet venomous? Enter 'y' or 'n': ").lower()
            if venomous != 'y' or venomous != 'n':
                raise ValueError
        except ValueError:
            print('Enter y or n!')
        else:
            err = False
    
    return (limbCount,venomous)

def get_fish_input():
    err = True
    while err:
        try:
            scale = input("Enter the scale condition. Only 'Smooth' or 'Rough' is allowed! ").lower()
        except Exception as e:
            print(e)
        else:
            err = False
    err = True
    while err:
        try:
            length = float(input("Enter the length of the pet in inches: "))
            if length < 0:
                raise ValueError
        except ValueError:
            print("Enter Positive integer")
        except Exception as e:
            print("Enter valid input", e)
        else:
            err = False
    return (scale,length)
```

```python
def option_1():
    petSelection = 0
    while petSelection < 1 or petSelection > 3:
        try:
            petSelection = int(input("Type of Pet?(1-Mammal;2-Amphibian;3-Fish)"))
            if petSelection < 1 or petSelection > 3:
                raise ValueError
        except ValueError:
            print('Wrong Selection! Enter valid input!')
            continue
    pName, pDOB, pWeight = get_animal_input()
    pOwnerName, pOwnerAddress = get_owner_input()
    owner = Owner(pOwnerName, pOwnerAddress)
    if petSelection == 1:
        mLitterSize, mHasClaws = get_mammal_input()
        animal = Mammal(pName, pDOB, pWeight, owner, mLitterSize, mHasClaws)
    elif petSelection == 2:
        aNumberOfLimbs, aIsVenomous = get_amphibian_input()
        animal = Amphibian(pName, pDOB, pWeight, owner, aNumberOfLimbs, aIsVenomous)
    elif petSelection == 3:
        fScaleCondition, fLength = get_fish_input()
        animal = Fish(pName, pDOB, pWeight, owner, fScaleCondition, fLength)
    return animal
```

```python
def option_2(pet_list):
    print("===================================================")
    print("Data of all Mammals in the database")
    print("===================================================")
    print("Pet Name\tDOB\tBirth Weight\tOwner Name\tOwner Address\tLitter Sizes\tHas Claws")
    for pet in pet_list:
        if isinstance(pet,Mammal):
            print(f'{pet.petName}\t\t{pet.dateOfBirth}\t{pet.birthWeight}\t{pet.owner.petOwner}\t{pet.owner.petOwnerAddress}\t{pet.litterSize}\t{pet.hasClaws}')
    
    print("===================================================")
    print("Data of all Amphibians in the database")
    print("===================================================")
    print("Pet Name\tDOB\tBirth Weight\tOwner Name\tOwner Address\tNumber Of Limbs\tIs Venomous")
    for pet in pet_list:
        if isinstance(pet,Amphibian):
            print(f'{pet.petName}\t\t{pet.dateOfBirth}\t{pet.birthWeight}\t{pet.owner.petOwner}\t{pet.owner.petOwnerAddress}\t{pet.numberOfLimbs}\t{pet.isVenomous}')
    print("===================================================")
    print("Data of all Fish in the database")
    print("===================================================")
    print("Pet Name\tDOB\tBirth Weight\tOwner Name\tOwner Address\tScale Condition\tLength")
    for pet in pet_list:
        if isinstance(pet,Fish):
            print(f'{pet.petName}\t\t{pet.dateOfBirth}\t{pet.birthWeight}\t{pet.owner.petOwner}\t{pet.owner.petOwnerAddress}\t{pet.scaleCondition}\t{pet.length}')
       
```

```python
def option_3(pet_list):
    print("=========================================")
    print("Pet Name and Current Weight of all pets")
    print("=========================================")
    for pet in pet_list:
        print(pet.computeCurrentWeight())
```

```python
def option_4(pet_list):
    owner_list = []
    for i in range(len(pet_list)):
        owner_list.append(pet_list[i].owner)
    
    owner_pet_count = {}
    owner_present = []
    
    for pet in pet_list:
        owner_name = pet.owner.petOwner
        if owner_name in owner_pet_count:
            owner_pet_count[owner_name] += 1
        else:
            owner_pet_count[owner_name] = 1
            
    print("================================================================")
    print("Owner Name  -  Pet Name  -  Owner address for multi-pet owners")
    print("================================================================")
    for owner in owner_list:
        owner_name = owner.petOwner
        if owner_name in owner_pet_count and owner_pet_count[owner_name] > 1 and owner_name not in owner_present:
            owner_present.append(owner_name)
            print(owner,end='\t')
            print("The Pet names are:")
            for pet in pet_list:
                if pet.owner.petOwner == owner_name:
                    print(f"\t{pet.petName}\t")
            
```

```python
def option_5(pet_list, curr):
    print("You chose to exit the program")
    yes_no = input("Are you sure (Y/N)?").lower()
    if yes_no.lower() == 'y':
        file = open('petdata.dat','ab')
        for obj in pet_list[curr:]:
            pickle.dump(obj,file)
        file.close()
        return True
    return False
    
```

```python
def menu():
    pet_list = read_file()
    selection = 1
    curr_count = len(pet_list)
    while selection >= 1 and selection <= 5:
        print("==== Menu ====")
        print("1. To add a pet\n2. To print pet information for all pet in the database\n3. To print pet name and current weight of all pets in the database\n4. To print the pet name, owner name and owner address for all owners with more than one pet\n5. To exit program")
        try:
            selection = int(input("Your selection: "))
            if selection < 1 or selection > 5:
                raise ValueError
        except ValueError:
            print('You must enter an integer between 1 and 5!')
            print('Try again')
            selection = 1
            continue
        if selection == 1:
            obj = option_1()
            pet_list.append(obj)
            print("Pet added to database")
        elif selection == 2:
            if len(pet_list) == 0:
                print('Not pets to print')
            else:
                option_2(pet_list)
        elif selection == 3:
            if len(pet_list) == 0:
                print('Not pets to print')
            else:
                option_3(pet_list)
        elif selection == 4:
            if len(pet_list) == 0:
                print('Not pets to print')
            else:
                option_4(pet_list)
        elif selection == 5:
            if option_5(pet_list, curr_count):
                break
        
    print("Exiting the program! Goodbye!")

        
```
